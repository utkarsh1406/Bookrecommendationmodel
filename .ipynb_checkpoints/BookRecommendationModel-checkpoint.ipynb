## Popularity Based Recommendation
##### We will show top 50 books with highest ratings (minimum voting critera = 50 votes)

```python
# Merge the users and rating
book_ratings = ratings.merge(books,on='ISBN')
```

```python
book_ratings.shape
```

```python
no_book_rating = book_ratings.groupby('Book-Title').count()['Book-Rating'].reset_index().rename(columns={'Book-Rating': 'noOfRatings'})
no_book_rating
```

```python
mean_book_rating = book_ratings.groupby('Book-Title')['Book-Rating'].mean().reset_index().rename(columns={'Book-Rating' : 'AverageRating'})
mean_book_rating
```

```python
no_book_rating = no_book_rating.merge(mean_book_rating, on='Book-Title')
no_book_rating.head()
```

```python
# finding top rated book with minimum 250 rating counts
popular_books = no_book_rating[no_book_rating['noOfRatings'] >= 250].sort_values(by='AverageRating', ascending=False).reset_index().head(50)
popular_books
```

```python
popular_df = popular_books.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title', 'noOfRatings', 'AverageRating','Book-Author','Year-Of-Publication','Publisher','Image-URL-M' ]]
```

```python
popular_df.head( 40)
```

```python
book_ratings.head(2)
```

```python
# getting users who have given atleast 100 books
x = book_ratings.groupby('User-ID').count()['Book-Rating'] > 100
intellect_user_ratings = book_ratings[book_ratings['User-ID'].isin(x[x].index)]
```

```python
# Book based filtering
y = intellect_user_ratings.groupby('Book-Title').count()['Book-Rating'] >= 50
intellect_user_ratings = intellect_user_ratings[intellect_user_ratings['Book-Title'].isin(y[y].index)]
```

```python
pt = intellect_user_ratings.pivot_table(index='Book-Title', columns='User-ID',values='Book-Rating')
```

```python
print(pt.head(3))
pt.fillna(0,inplace=True)
pt.head(5)
```

```python
pt.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_score = cosine_similarity(pt)
```

```python
similarity_score.shape
```

- First fetch index using book name

```python
import numpy as np 
np.where(pt.index == '24 Hours')[0]
```

```python
def Book_Suggestion(book_name:str)-> list[str]:
    # Index Fetching
    suggestion =[]
    index = np.where(pt.index == book_name)[0][0]
    similar_books = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    for each in similar_books:
        items = []
        temp_df = books[books['Book-Title'] == pt.index[each[0]]]
        items.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        items.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        items.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-S'].values))
        items.extend(list(temp_df.drop_duplicates('Book-Title')['Publisher'].values))
        suggestion.append(items)
    return suggestion    
```

```python
Book_Suggestion('The Color Purple')
```

```python
import pickle
pickle.dump(popular_df,open('popular.pkl','wb'))
pickle.dump(books,open('books.pkl','wb'))
pickle.dump(similarity_score,open('similarity_score.pkl','wb'))
pickle.dump(pt,open('pt.pkl','wb'))
```
